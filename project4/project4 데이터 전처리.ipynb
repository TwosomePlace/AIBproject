```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/bgg-19m-reviews.csv')
```

```python
# 기본 데이터
df
```

```python
# 불필요한 열 제거
df.drop(['Unnamed: 0','user'],axis = 1, inplace = True)
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 리뷰(comment) 결측치 제거
df.dropna(inplace=True)
df
```

```python
# 리뷰 수에 따라 정렬
review_count = df.groupby(['ID']).size().reset_index(name='counts').sort_values('counts',ascending = False)
review_count
```

```python
# 대중성을 위해 리뷰 수가 1500개(최다 리뷰의 10% 정도) 이상인 게임 목록 추출
ID = review_count[review_count['counts']>1500]['ID'].unique()
ID
```

```python
# 목록에 포함된 게임의 데이터만 추출
df = df[df['ID'].isin(ID)]
df
```

```python
# 리뷰 길이가 100(평균의 50% 정도) 이상인 데이터 추출
df = df[df['comment'].str.strip().map(len)>=100]
df
```

```python
# 인덱스 초기화
df.reset_index(drop=True,inplace=True)
```

```python
# 파일 저장
df.to_csv('/content/drive/MyDrive/project4(boardgame_review).csv', index=False)
```
